Plot function and add noise

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.stats import norm
import random

x = np.linspace(0,5,20)

y = 2*np.sin(1.5*x)

# plot the function
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel("f(x) = 2sin(1.5x)")
# show the plot
#plt.show()
```

Add Noise

```python
sample_x = np.random.rand(20)*5
sample_y = 2 * np.sin(1.5*sample_x) + np.random.rand(20)
plt.plot(sample_x,sample_y,'d')
```

```python
plt.plot(sample_x,sample_y,'d')
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('f(x)')
```

Using numpy.polyfit to generate the models.

```python
p2 = np.poly1d(np.polyfit(sample_x,sample_y,2))
p3 = np.poly1d(np.polyfit(sample_x,sample_y,3))
p4 = np.poly1d(np.polyfit(sample_x,sample_y,4))
p5 = np.poly1d(np.polyfit(sample_x,sample_y,5))
p6 = np.poly1d(np.polyfit(sample_x,sample_y,6))
p7 = np.poly1d(np.polyfit(sample_x,sample_y,7))
p8 = np.poly1d(np.polyfit(sample_x,sample_y,8))
p9 = np.poly1d(np.polyfit(sample_x,sample_y,9))

plt.plot(sample_x,sample_y,'d',x,p2(x),'--')
plt.plot(x,p3(x),'--')
plt.plot(x,p4(x),'--')
plt.plot(x,p5(x),'--')
plt.plot(x,p6(x),'--')
plt.plot(x,p7(x),'--')
plt.plot(x,p8(x),'--')
plt.plot(x,p9(x),'--')
#plt.plot(sample_x,sample_y,'d',f_x, p9(f_x))
plt.xlabel('x')
plt.ylabel('f(x) = 2sin(1.5x)')
plt.show()
```

Question 1: Produce a plot of error as a function of model order up to 9, similar to that shown in Fig.4.7 of the Alpaydin book

```python
polmodels = [p2,p3,p4,p5,p6,p7,p8,p9]
def get_error(p):
  error_array=[]
  for i in np.arange(0,5,0.25):
    true = 2*np.sin(1.5*i)
    predict = p(i)
    error_array.append(true-predict)
  plt.plot(x,error_array, '--')

for model in polmodels:
  get_error(model)
plt.ylabel('Error')
plt.xlabel('x')
plt.show()
```

Observe the effect of overfitting as the model complexity increased while you
trained the models: Overfitting causes the model to skew high or low. Increasing the variance means that the model has a larger body of training data and can better predict its cooefficents.
